```python
class Bank:
    b_name = "Kumari bank"
    location = "Kathmandu"
    
    def __init__(self,  person_name, account_number, balance=0):
        self.balance = balance
        self.person_name = person_name
        self.account_number = account_number
        #self.email = f"{self.person_name}@gmail.com" 
        #When we use this if object's name is changed in future but email will not be changed
        
        
        #To fix the issue we use the property
        @property #After using these, method will be converted into attribute 
        def email(self):
            return(f"{self.person_name}@gmail.com") #It will be method if not used property decorator 
        @email.setter
        #def email(self, new_mail):
            #self.email = new_mail #it will return the recursive error
        def email(self, new_name):
            self.name = new_name
             
           
   
        

        
        
    #init method
    
    #deposit
    def deposit(self,deposit):
        print(f"Your current balance is: {self.balance}")
        self.total = self.balance + deposit
        print(f"Balance after deposit is: {self.total}")
     #Withdraw
    def withdraw(self, withdraw):
        print(f"Your current balance is: {self.balance}")
        
        
        if self.balance < withdraw:
            print(f"Insufficient balance")
            print(f"Available balance: {self.balance}")
        else:
            self.total = self.balance - withdraw
            print(f"Balance after withdraw is: {self.total}")
             
        
        
    @classmethod #Decorator is used 
    def change_location(cls,new_location): #It will change the function permanently
        cls.location = new_location
    
    #Display
    def display(self):
        print(f"Location is changed to {self.location}")
        
    
         
```

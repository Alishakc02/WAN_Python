Steps of becoming object:
- Creation
- Initialization
  

__new__() - Creating a new instance of a class.
          - It allocates the memory and returns the new object 
          - Rarely overridden
          - Singleton design pattern, immutable objects 


__init__() - It initializes the newly instance of the class.
           - It is used as constructor 
           - Commonly overridden
           - Set attribute values 
  

```python
class MarkerIndustry:
    def __new__(cls):
        print("Creating object")
        return super().__new__(cls)
    
    def __init__(self):
        print("Initializing object")
        
```

```python
m1 = MarkerIndustry()
print(m1)
```

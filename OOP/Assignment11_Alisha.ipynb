1. Create a Rectangle class with: 
●  Attributes: length, width
●  Methods:
○  area() 
○  perimeter()
-   Create 3 Rectangle objects. One object displays area() and the other 
object displays perimeter()

```python
class Rectangle():
    length = 10
    width = 20
    
    #METHOD DEFINE
    def area(self):
        return self.length * self.width
    
    #For perimeter
    def perimeter(self):
        return 2 * (self.length + self.width)

    

    
```

```python
#Object creation
rect_area = Rectangle()
print(f"The area of the rectangle is: {rect_area.area()}")

#For perimeter
p1 = Rectangle()
p2 = Rectangle()
print(f"The perimeter of the rectangle is: {p1.perimeter()}")
print(f"The perimeter of the rectangle is: {p2.perimeter()}")
```

2. Create a Person class with: 
●  Attributes: name, age
●  Methods:
○  introduce() → prints "Hi, I am [name], and I am [age] years old." 
Create 2 Person objects and call introduce() on both.

```python
class Person():
    name = "Alisha"
    age = 78
    
    #Method define
    def display(self):
        return f"Name: {self.name}, Age: {self.age}"
```

```python
#Object creation
p1 = Person()
p2 = Person()
print(p1.display())
print(p2.display())
```

4. Create a Circle class with: 
●  Attribute: radius
●  Methods:
○  area() → returns area (use 3.14 for π) 
○  circumference() → returns circumference
Create one circle with radius 5 and print area and circumference.

```python
class Circle():
    radius = 5
    #Methods
    
    def area(self):
        print(f"The area of the circle is: {3.14 * self.radius ** 2}") 
    
    def circumference(self):
        print(f"The circumference of the circle is: {2 * 3.14 * self.radius}")
```

```python
#Object creation
c1 = Circle()
c1.area()
c1.circumference()
```

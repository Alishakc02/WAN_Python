Create a class BankAccount with attributes account_number and balance. Include methods 
deposit(amount) and withdraw(amount).

```python
class BankAccount:
    account_number = "1234567890"
    balance = 50000
    
    def deposit(self, amount):
        print(f"The current balance is: {self.balance}")
        print(f"The amount after deposit is: {self.balance + amount}")
    
    def withdraw(self, amount):
        print(f"The current balance is: {self.balance}")
        if amount > self.balance:
            print(f"Insufficient balance in the account")
        else:
            print(f"The amount after deposit is: {self.balance - amount}")
```

```python
b1 = BankAccount()
b1.deposit(8000)
```

```python
b1.withdraw(3000)
```

Write a class Circle with attribute radius. Include methods to calculate area and 
circumference. (Use 3.14 for π)

```python
class Circle:
    radius = 6
    def area(self):
        print(f"The area of the circle is: {3.14 * self.radius * self.radius}")
    #Circumference of the circle
    def circumference(self):
        print(f"The circumference of the circle is: {2 * 3.14 * self.radius}")
```

10.  Create a class Counter with an attribute count. Include methods increment(), decrement(), 
and reset().

```python
class Counter:
    count = 0
    #At first initialize the value of count to 0

    def increment(self, number):
        self.count += number
        print(f"The value of count after increment is: {self.count}")
        #Increment the count value

    def decrement(self, number):
        if self.count - number < 0: #Checking condition
            self.count = 0
        else:
            self.count -= number
        print(f"The value of count after decrement is: {self.count}")

    def reset(self):
        self.count = 0
        print(f"The value of count after reset is: {self.count}")
```

```python
count1 = Counter()
count1.increment(10)
```

```python
count1.decrement(11)
```

```python
count1.increment(10)
```

```python
count1.reset()
```

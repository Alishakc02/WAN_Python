Write a fuction display_employee(**kwargs) that accepts employee information using keyword arguments and displays each key-value pair.

Take at least 5 employee information by yourself.

````python

def display_employee(**employees):
  try:
    if "name" not in employees:
            raise KeyError("Employee name is required.") #There is no key called name so raised error
    for key, value in employees.items():
      print(f"{key}: {value}")
  except Exception as e:
    print(f"An error occurred: {e}")
  finally:
    print("Calculation completed.")
````

````python


display_employee(

    age='90',
    position="Content Creator",
    department="Marketing",
    salary=50000
)
````

Question 1: Student Grade Calculator Write a Python program that calculates a student's total marks, percentage, and final letter grade based on their scores in Math, Science, English, History, and Art. Requirements:

Prompt the user to enter the marks for all five subjects individually.
Convert each input into an integer.
Calculate the total marks obtained (out of 500).
Calculate the overall percentage.
Assign a grade: A (90-100), B (80-89.99), C (70-79.99), D (60-69.99), F (below 60).
Display Total Marks, Percentage (2 decimal places), and Final Grade.**

````python
def student_grade(math, science, english, history, art):
    try:
        total_marks = math + science + english + history + art
        percentage = (total_marks / 500) * 100

        if percentage >= 90:
            grade = "Grade: A"
        elif percentage >= 80:
            grade = "Grade: B"
        elif percentage >= 70:
            grade = "Grade: C"
        elif percentage >= 60:
            grade = "Grade: D"
        else:
            grade = "Grade: F"

        return total_marks, percentage, grade

    except Exception as e:
        print(f"An error occurred: {e}")

    finally:
        print("Grade calculation completed.")



````

````python
try:
    math = int(input("Enter the marks for Math: "))
    science = int(input("Enter the marks for Science: "))
    english = int(input("Enter the marks for English: "))
    history = int(input("Enter the marks for History: "))
    art = int(input("Enter the marks for Art: "))

    result = student_grade(math, science, english, history, art)

    if result:
        total_marks, percentage, grade = result
        print(f"Total Marks: {total_marks}")
        print(f"Percentage: {percentage:.2f}%")
        print(grade)

except ValueError:
    print("Please enter valid integer values.")

except Exception as e:
    print(f"An error occurred: {e}")
````

**Write a Python program to print all even numbers between 1 and 100.**

````python

def even_number():
    try:
        n = int(input("Enter the limit: "))

        for i in range(1, n + 1):
            if i % 2 == 0:
                print(i)

    except ValueError:
        print("Please enter a valid integer.")

    except Exception as e:
        print(f"An error occurred: {e}")

    finally:
        print("Calculation completed.")

even_number()
````

Question 3: FizzBuzz Write a Python program that prints numbers from 1 to 100.

Print "Fizz" for multiples of 3.
Print "Buzz" for multiples of 5.
Print "FizzBuzz" for multiples of both.
Otherwise, print the number.

````python
def fizz():
    try:

        for i in range(1, 101):
            if i % 3 == 0 and i % 5 == 0:
                print(f"FizzBuzz: {i}")
            elif i % 0 == 0: #Error will occur here
                print(f"Fizz: {i}")
            elif i % 5 == 0:
                print(f"Buzz: {i}")
            else:
                print(i)

    except Exception as e:
        print(f"An error occurred: {e}")

    finally:
        print("Program completed.")

fizz()
````

Question 4: Multiplication Table Write a Python program that asks the user to enter a number and prints its multiplication table from 1 to 10, except for even numbers, using a loop.

````python
def multiplication_table(n):
    try:
        i = 1
        while i <= 10:
            if i % 2 == 0:
                i += 1
                continue
            print(f"{n} * {i} = {n * i}")
            i += 1

    except Exception as e:
        print(f"An error occurred: {e}")

    finally:
        print("Program completed.")
````

````python
try:
    n = int(input("Enter a number: "))
    multiplication_table(n)

except ValueError:
    print("Please enter a valid integer.")
````

**Write a Python program that stores the secret number as 25.

Ask the user to guess the number.
Keep asking until the correct number is entered.
If the guess is too low, print "Too Low!"
If the guess is too high, print "Too High!"
When the user guesses correctly, print "Congratulations! You guessed the correct number."**

````python
def guess_number():
    secret = 25

    while True:
        try:
            guess = int(input("Guess the number: "))

            if guess == secret:
                print("Congratulations! You guessed the correct number.")
                break
            elif guess < secret:
                print("Too Low!")
            else:
                print("Too High!")

        except ValueError:
            print("Please enter a valid integer.")



        except EOFError:
            print("\nNo input received.")
            break

        except Exception as e:
            print(f"An error occurred: {e}")
            break

    print("Game Over.")

guess_number()
````

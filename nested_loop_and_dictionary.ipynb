**Nested Dictionary**

```python
#Dictionary define

dict1 ={"first":{"name":"Alisha", "roll":90, "sub":["Math","C"]},
        "second":
       {"place":"Bkt"},
        "third":
       {"food":"pizza", "num":6.5}
       }
print(dict1)
```

```python
print(dict1['first']['roll'])
```

```python
print(dict1['first']['sub'])
```

```python
print(dict1['second']['place'])
```

```python
#Modifying
dict1['first']['roll']=45
print(dict1)
```

```python
dict1['first']['sub'][0]="Java"
print(dict1)
```

```python
dict1['fourth']={"class":"Boring"}
print(dict1)
```

```python
#Remove

dict1.pop('fourth')
print(dict1)
```

**Dictionary methods**

```python
print(dict1.keys())
```

```python
print(dict1.values())
```

```python
print(dict1.items())
```

```python
#Update
dict1['first'].update({"name":"Ankit"})
print(dict1)
```

```python
#get
print(dict1.get('first'))
```

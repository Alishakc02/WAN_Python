```python
#Set joints
#Union
s1={1,2,3,4}
s2={2,4,5}
s3=s1.union(s2)
print(s3)
print(s1)
print(s2)
```

```python
#update

s1={1,2,3,4}
s2={2,4,5}
s3=s1.update(s2)
print(s3)
print(s1) #update the set s1 instead of s3
print(s2)
```

```python
# using |
s1={1,2,3,4,5,6}
s2={2,3,9,0}
print(s1 | s2)
print(s1) 
print(s2)
```

```python
#intersection
s1={1,2,3,4,5,6}
s2={1,2,3,9,0}
s3=s1.intersection(s2)
print(s3)
```

```python
print(s1.intersection_update(s2))
s1
```

```python
print(s1 & s2) #using & sign 
```

```python
#difference
s1={8,9,0}
s2={3,4,9}
s3=s1.difference(s2)
print(s3)
```

```python
print(s1.difference_update(s2))
s1
```

```python
print(s1 - s2)
```

```python
#symmetric difference - ^ is used - elements except common element in the sets 
s1={1,2,3,4,5,6}
s2={0,1,3,5}
print(s1.symmetric_difference(s2))
```

```python
print(s1.symmetric_difference_update(s2))
print(s1)
```

```python
#using _ ^ sign
s1 ={9,0}
s2={8,0}
print(s1 ^ s2)
```

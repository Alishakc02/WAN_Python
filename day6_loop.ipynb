```python
x=[100,101,200,250,300,310]
for i in x:
    if i%100==0:
        print(f"{i} is divisible by 100")
    else:
        print(f"{i} is not divisible by 100")
```

```python
#continue
x=[100,101,200,250,300,310]
for i in x:
    if i%100==0:
        if i == 200:
            continue
        print(f"{i} is divisible by 100")
        
    else:
        print(f"{i} is not divisible by 100")
```

```python
#break
x=[100,101,200,250,300,310]
for i in x:
    if i%100==0:
        if i == 200:
            break
        print(f"{i} is divisible by 100")
        
    else:
        print(f"{i} is not divisible by 100")
```

Nested For Loop

```python
x=[1,2,3,4,3,2,2,1]

for i in x:
    print(f"Numbers:{i}")
    
    for j in x:
        sum=i+j
        
        if sum == 5:
            print(f"{i} and {j} makes {sum}")
    
```

```python
x=[1,2,3,4,3,2,2,1]

for i in x:
    print(f"Numbers:{i}")#i represents the item instead of index in for loop
```

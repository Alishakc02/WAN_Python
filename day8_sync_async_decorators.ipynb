Synchronous and Asynchronous function

```python
import time
```

Function in the queue in the petrol pump in synchronous manner
- Use def only to  define synchronous function
- Use main function to call the function
- We can call both synchronous and asynchronous function in synchronous function
  

```python
def petrol_pump(v_id):
    print(f"Turn of {v_id}")
    time.sleep(2) #to open lead of tank
    print(f"Filling petrol: {v_id}")
    time.sleep(5) #to pay cash
    print(f"Paid cash: {v_id}")
    
    
    
```

```python
def main():
    start_time = time.time()
    #It counts the starting time
    
    #Vehicle number
    petrol_pump(1023)
    petrol_pump(1480)
    end_time = time.time()
    #it stops the time 
    print(f"Total execution time: {end_time - start_time:.2f} seconds")
    
    
    
```

```python
main()
```

Function in the queue in the petrol pump in asynchronous manner
- Use aync keyword to define aynchronous function
- Use await function_name to call the function
- We can call only the aynchronous function
- API Call, Reduce latency(delay)
  

```python
import asyncio
```

```python
async def petrol_pump1(v_id):
    print(f"Turn of: {v_id}")
    await asyncio.sleep(2)
    print(f"Filling petrol: {v_id}")
    await asyncio.sleep(5)
    print(f"Paid cash: {v_id}")
    
```

```python
async def main():
    start_time = time.time()
    await asyncio.gather(petrol_pump1(1023), petrol_pump1(1480))
    end_time = time.time()
    print(f"Total execution time: {end_time - start_time:.2f}Seconds" )
    
    
```

```python
await main()
```

Decorators

Definition:
           - Function that contain function and return function
           - Enhance the behaviour of original function without modifying function
           - logging, authentication, caching

```python
def decorator_func(main): #It will take function as a parameter and returns function
    def inner():
        main()
        print('World')
    return inner
    
```

```python
@decorator_func #Use decorator_name at the beginning to initialize or write the name of the decorator 
def main():
    print("Hello")
```

```python
main()
```

```python
def decorator_sum(calc_sum): #Include funcrtion over here as calc_sum is a function 
    def inner(a,b): #It should include as much parameter as calc_sum has
        a+=5 #Increase value of a by 5 so that in another function it will perform addition after its value is increased
        sum = calc_sum(a,b)
        return sum
    return inner

        
```

```python
@decorator_sum #calling decorator
def calc_sum(a,b):
    return a + b

```

```python
calc_sum(2,3)
```

*args vs **kwargs
*args - Accepts any number of positional arguments
       - Stored as tuple
       - info(4,6)
**kwargs - Accepts any number of keyword arguments 
         - Stored as dictionary
         - info(a=4, b=9)


```python
def user_info(*user_name):#we can pass any number of arguments over here and it will return tuple
    print(type(user_name))
    for i in user_name:#To print separately we use for loop over here 
        #in - membership operator for tuple 
        

        print(f"User name: {i}")
    
```

```python
user = user_info("Alisha","Samana","Kritisha")
user
```

Function for the set of tuples

```python
def user_info(*info):
    for name, id in info:
        #print(type(info))
        #print(f"User name: {name} ")
        #print(f"User id: {id}")
        print(f"User name: {name}, User id: {id} ")
    
```

**kwargs

```python
def user_info(**info):
    print(info)
    print(type(info))
    print(f"Username: {info['name']}, Age: {info['age']}, College: {info['college']}")
   
    for k, v in info.items():
        print(f"{k} : {v}")
        
        
    
```

```python
a=user_info(name= "Deepal", age= 23, college="NCIT")
a
```

GLOBAL VARIABLE

```python
x = 5
def scope():
    global a #global keyword 
    a = 4
    print(f"Inside function x", x)
    print(f"Inside function a", a)
print(f"Outside a", a)
print(f"Outside x", x)
```

```python
scope()
```

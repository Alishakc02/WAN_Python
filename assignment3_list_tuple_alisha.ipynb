Take a list, reverse it completely using slicing, and then swap its new first and
last elements.

```python
list1=[1,2,3,5,6,"Alisha"]
#reverse using slicing
rev=list1[::-1]
print(rev)
```

```python
#swap the first and last
rev[0],rev[5]=rev[5],rev[0]
print(rev)
```

Given a = [1, 2, 3, 5, 6, 7, 8, 9] and b = [1, 3, 5, 7, 9], return a new
list containing only the elements that are unique across both lists (items present
in one list but not both)

```python
a = [1, 2, 3, 5, 6, 7, 8, 9]
b = [1, 3, 5, 7, 9]
result= list(set(a).symmetric_difference(set(b)))
print(result)
```

. Given the tuple nums = (1, 5, 3, 5, 2, 5, 4), find exactly how many times 5
appears and the index at which it first appears.

```python
nums = (1, 5, 3, 5, 2, 5, 4)
print(nums.count(5))
print(nums.index(5))
```

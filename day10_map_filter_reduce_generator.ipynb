MAP, FILTER, REDUCE FUNCTION
Syntax:
       map(function, iterable)

- Returns map iterator
- filter(function, iterable)
- reduce(function, iterable)

```python
#map
l=[1,2,3,4,5,6,7]
a = map(lambda x: x>5, l)
print(list(a))
```

```python
#filter
l=[1,2,3,4,5,6,7]
a = filter(lambda x: x>5, l)
print(list(a))
```

```python
import functools
```

```python
#reduce 
#It requires functools to use reduce 
l=[1,2,3,4,5]
a = functools.reduce(lambda x, y: x+y, l)
print(a)
```

```python
def func2(n):
    for i in range(n):
        if i%2 == 0:
            yield i
            print("Hi")
```

```python
#Use next with yield 
a= func2(1000000)
print(next(a)) #It checks for the first value 0
```

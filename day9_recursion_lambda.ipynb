RECURSION FUNCTION

Condition:
- Function must call itself
- There should be a terminating condition
  


Factorial of the number using a function

```python
def factorial(n):
    
    fact = 1
    for i in range(1, n+1):
        fact = fact * i
        
    return fact
 
```

```python
a = factorial(4)
a
```

```python
#Recursion

def text_rec(text):
    if len(text) == 0:
        print("Done")
    else:
        print(text)
        return text_rec(text[:-1]) #Recursion call
text_rec("Hello")
```

```python
#Recursive function to calculate multiplication table
def multiply(a,b):
    if b == 1:
        return a
    else:
        return a + multiply(a,b-1)
multiply(2,3)
```

LAMBDA 
Definition:
- A small anonymous function defined without a formal name
- A lambda function can take any number of arguments, but can only have one expression
- Use lambda as keyword


SYNTAX:
   lambda argument:expression

USES:
- Higher order function(input-function)
  

```python
#Lambda function
b = lambda x: x**2
b(2)
```

```python
print(type(b))
```

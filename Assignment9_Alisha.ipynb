1. Use map() and lambda to double all numbers in this list: [1, 2, 3, 
4, 5].

```python
#define list
li = [1,2,3,4,5] 
#lambda function to double the value of each element in the list
x = lambda y: 2*y
#mapping function
a = list(map(x,li))
#result display
a
```

2. Use map() and lambda to convert this list of strings to 
uppercase: ["apple", "banana", "cherry"].

```python
list1 = ['apple', 'banana', 'cherry']
x = lambda y:y.upper()
a = list(map(x,list1))
a
```

3. Write a recursive function to find the factorial of a number and 
trace the output in a copy as well.

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1) #Recursive call to the function
factorial(5) #Result display
```

4. Write a recursive function to find the length of a string and trace 
the output as well.

```python
def str_len(text):
    if text == " ":
        return 0
    else:
        return str_len(text[1:]) + 1 #Recursive call to the function
str_len("Hello World") #Result display
```

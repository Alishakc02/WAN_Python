ERROR
- An error is a mistake or fault in the program that causes the program to fail or behave unexpectedly 


Types:
1). Compile time error - Syntax and semantic(meaning) error
2). Run-time error - Exceptions 
3). Logical error

Exception:
- An exception is a runtime event that interrupts the normal flow of the program

Uses of exception handling:
- Prevents program crashes
- Sudden termination of program cause data loss from database from a file
- Makes debugging easier

Syntax:
  try: 
     Risky_code
  except:
      Handle_error
  finally:
       Final statement

```python
#try-except exception handling

try:
    l1 = [1,2,4,5]
    print(l1[6])
except Exception as e: #hierarchy name that contains the error 
    print(type(e).__name__)
    print(e)
finally:
    print("Finally executed") 

```

```python
#TRY-EXCEPT WITHOUT ny error 
try:
    l1 = [1,2,4,5]
    print(l1[0])
except Exception as e: #hierarchy name that contains the error 
    print(type(e).__name__)
    print(e)
finally:
    print("Finally executed") #even if error occurs or not finally block will be executed 
```

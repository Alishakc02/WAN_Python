### Python TypeCasting

| Variable | `int()` | `float()` | `complex()` | `str()` |
| :--- | :---: | :---: | :---: | :---: |
| Integer | Yes | Yes  | Yes | Yes |
| Float | Yes | Yes | Yes | Yes |
| Complex | No | No | Yes | Yes |
| String | No | No | No | Yes |



```python
a = 5
print(type(a))
```

```python
b = float(a)
print(b)
print(type(b))
```

```python
c = complex(a)
print(c)
print(type(c))
```

```python
s = str(a)
print(s)
print(type(s))
```

```python
c = 1+1j
print(str(c))
print(type(str(c)))
```

```python
name = input("Enter your Name: ")
age = input('Enter your Age:')
print(type(name),type(age))
print("User Name: ",name,"User Age:",age )
print(f"User name is {name} and age is {age}")
```
